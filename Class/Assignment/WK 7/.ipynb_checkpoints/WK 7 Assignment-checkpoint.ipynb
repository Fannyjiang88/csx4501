```python
from sklearn import datasets
import pandas as pd
df = datasets.load_diabetes(as_frame=True) 
features_df = df.data
target_df = df.target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
#Split the data into a training set and test set
X_train, X_test, y_train, y_test = train_test_split(features_df, target_df, test_size =.2)
```

```python
linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)
```

```python
linear_regression.score(X_test,y_test)
```

```python
linear_regression.coef_
```

```python
linear_regression.intercept_
```

```python
#Plot the linear fit together with the training data
plt.plot(X_train, y_train, 'b-',X_test,y_test,'ro')
```

```python
#Use the model to generate predictions on the test data and then compare against the actual
#target values by calculating the mean squared error and R-squared values.
from sklearn.metrics import r2_score, mean_squared_error
```

```python
y_pred = linear_regression.predict(X_test)
```

```python
r2score = r2_score(y_test, y_pred)
msescore = mean_squared_error(y_test, y_pred)
```

```python
print("Testing score R2 : ", r2score)
print("Testing score StdDev : ", np.sqrt(msescore))
```

```python
theta_0 = linear_regression.coef_
theta_0
```

```python
intercept = linear_regression.intercept_
intercept
```

```python
plt.plot(y_pred, label="Prediction")
plt.plot(y_test.values, label="Actual")
plt.legend()
```
