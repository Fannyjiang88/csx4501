```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
h2=df.data[['s1','s2']].copy()
```

```python
h2.sample(10)
```

```python
X = h2['s1']
y = h2['s2']
```

```python
X.head()
y.head()
```

```python
#Split the data into a training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =.3)
```

```python
#Using this feature and target, perform linear regression with sklearn's LinearRegression
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
linear_regression = LinearRegression()
model = linear_regression.fit(X_train, y_train)
```

```python
#Print the coefficients of the model
y_pred = model.predict(X_test)
```

```python
df = pd.DataFrame({'test': y_test, 'predicted': y_pred})
```

```python
df.sample(10)
```

```python
print("Training score : ", linear_regression.score(X_train, y_train))
```

```python
#Plot the linear fit together with the training data
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, c='r')
```

```python
#Use the model to generate predictions on the test data and then compare against the actual
#target values by calculating the mean squared error and R-squared values.
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2score = r2_score(y_test, y_pred)
msescore = mean_squared_error(y_test, y_pred)
```

```python
print("Testing score R2 : ", r2score)
print("Testing score StdDev : ", np.sqrt(msescore))
```

```python
theta_0 = linear_regression.coef_
theta_0
```

```python
intercept = linear_regression.intercept_
intercept
```

```python
plt.plot(y_pred, label="Prediction")
plt.plot(y_test.values, label="Actual")
plt.legend()
```
